```python
# import libraries
import pandas as pd

# read dataset
df = pd.read_csv('data/who_suicide_statistics.csv')
```

## Descriptive Statistics

```python
# Calculate descriptive statistics for numerical columns
descriptive_stats = df[['suicides_no', 'population']].describe().T

# Add median to the descriptive statistics
descriptive_stats['median'] = df[['suicides_no', 'population']].median()

descriptive_stats
```

## Distribution Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visualisation style
sns.set(style="whitegrid")

# Distribution of Number of Suicides
plt.figure(figsize=(12, 6))
sns.histplot(df['suicides_no'], bins=50, kde=True, color='blue')
plt.title('Distribution of Number of Suicides')
plt.xlabel('Number of Suicides')
plt.ylabel('Frequency')
plt.show()

# Distribution of Population
plt.figure(figsize=(12, 6))
sns.histplot(df['population'], bins=50, kde=True, color='green')
plt.title('Distribution of Population')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()

# Distribution of Suicides by Gender
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='sex')
plt.title('Distribution of Suicides by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Distribution of Suicides by Age Group
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='age')
plt.title('Distribution of Suicides by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Distribution of Suicides by Year
plt.figure(figsize=(12, 6))
sns.histplot(df['year'], bins=50, kde=True, color='purple')
plt.title('Distribution of Suicides by Year')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()

# Due to the large number of countries, we can show the top 10 countries with the most records
top_countries = df['country'].value_counts().head(10).index
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['country'].isin(top_countries)], y='country', order=top_countries)
plt.title('Top 10 Countries with Most Records')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

```python
# Compute correlation matrix
correlation_matrix = df[['suicides_no', 'population']].corr()

print(correlation_matrix)
```

## Categorical Analysis

```python
# Distribution of Suicides by Gender
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='sex', y='suicides_no', hue='sex', multiple='stack', palette='Set1')
plt.title('Distribution of Suicides by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Suicides')
plt.show()
```

```python
# Distribution of Suicides by Age Group
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='age', y='suicides_no', hue='age', palette='Set2')
plt.title('Distribution of Suicides by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Suicides')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of Suicides by Age Group
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='age', y='suicides_no')
plt.title('Distribution of Suicides by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Suicides')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of Suicides by Year
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='year', y='suicides_no', hue='year', multiple='stack', palette='Set4')
plt.title('Distribution of Suicides by Year')
plt.xlabel('Year')
plt.ylabel('Number of Suicides')
plt.show()
```
