```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read dataset
df = pd.read_csv('data/who_suicide_statistics.csv')
```

## 1. Suicide Rates by Age Group

```python
# Aggregate data by age group
age_group_trend = df.groupby('age')['suicides_no'].sum().reset_index()

# Plot suicide rates by age group
plt.figure(figsize=(12, 6))
sns.barplot(data=age_group_trend, x='age', y='suicides_no', palette='viridis', hue='age')
plt.title('Suicide Rates by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Suicides')
plt.xticks(rotation=45)
plt.show()
```

## 2. Suicide Rates by Gender

```python
# Aggregate data by gender
gender_trend = df.groupby('sex')['suicides_no'].sum().reset_index()

# Plot suicide rates by gender
plt.figure(figsize=(12, 6))
sns.barplot(data=gender_trend, x='sex', y='suicides_no', palette='viridis')
plt.title('Suicide Rates by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Suicides')
plt.show()
```

## 3. Suicide Rates by Age Group and Gender

```python
# Aggregate data by age group and gender
age_gender_trend = df.groupby(['age', 'sex'])['suicides_no'].sum().reset_index()

# Plot suicide rates by age group and gender
plt.figure(figsize=(12, 6))
sns.barplot(data=age_gender_trend, x='age', y='suicides_no', hue='sex', palette='viridis')
plt.title('Suicide Rates by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Number of Suicides')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```
