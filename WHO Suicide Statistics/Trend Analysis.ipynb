```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/who_suicide_statistics.csv')
```

## 1. Global Trend Analysis Over the Years

```python
# Aggregate data by year
global_trend = df.groupby('year')['suicides_no'].sum().reset_index()

# Plot global trend
plt.figure(figsize=(12, 6))
plt.plot(global_trend['year'], global_trend['suicides_no'], marker='o')
plt.title('Global Suicide Rates Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Suicides')
plt.grid(True)
plt.show()
```

## 2. Country-Specific Trend Analysis

```python
# List of specific countries to analyze
countries = ['Argentina', 'Japan', 'Italy', 'Brazil', 'Netherlands']

# Plot trends for specific countries
plt.figure(figsize=(12, 6))
for country in countries:
    country_trend = df[df['country'] == country].groupby('year')['suicides_no'].sum().reset_index()
    plt.plot(country_trend['year'], country_trend['suicides_no'], marker='o', label=country)

plt.title('Suicide Rates Over the Years by Country')
plt.xlabel('Year')
plt.ylabel('Number of Suicides')
plt.legend()
plt.grid(True)
plt.show()
```
